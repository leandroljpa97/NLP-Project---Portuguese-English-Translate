# Portuguese to English translater
## Data Cleaning, and split the dataset

<a> </a>
## 1.0) Imports

```python
import string 
import unidecode
import pandas as pd
import random
import math
```

<a></a>
## 1.1) Data Pre-processing and Data Cleaning

```python
file = open('por.txt', mode='rt', encoding='utf-8')
text = file.read()
# close the file
file.close()

# each line is separed by \n
lines = text.strip().split('\n')

dataSetAux = []
for item in lines:
    dataSetAux.append(item.split('\t'))

dataSetAux
```

```python
#As it can be observed, the 3rd position contains another information, that it is not needed. 
#We just want the first and the second positions of each array element of the list

#Besides, the punctuation has to be removed
#Set everything to lowerCase
# Remove any remaining tokens that are not alphabetic (accents in words)

dataSetPortuguese = []
dataSetEnglish = []

for i in dataSetAux:
    # remove punctuation
    portugueseWord = i[0].translate(str.maketrans('', '', string.punctuation))
    englishWord = i[1].translate(str.maketrans('','', string.punctuation))
    
    #normalize all words: remove all the accents
    portugueseWord = unidecode.unidecode(portugueseWord)
    englishWord = unidecode.unidecode(englishWord)
    
    #transform everything in lower case
    portugueseWord = portugueseWord.lower()
    englishWord = englishWord.lower()
    
    
    dataSetPortuguese.append(portugueseWord)
    dataSetEnglish.append(englishWord)



dataSetPortuguese
```

<a> </a>
## 1.2 Split the dataset in train and validation set

```python
#Split the dataset into training set and validation set and save in another file

sizeDataSet = len(dataSetPortuguese)
nrTrainSamples = math.floor(sizeDataSet*0.8)
nrValidSamples = sizeDataSet - nrTrainSamples

#use the same seed (4 in this case) to both lists
random.Random(4).shuffle(dataSetPortuguese)
random.Random(4).shuffle(dataSetEnglish)


dataSetPortugueseTrain, dataSetPortugueseValid = dataSetPortuguese[:nrValidSamples], dataSetPortuguese[nrValidSamples:]
dataSetEnglishTrain, dataSetEnglishValid = dataSetEnglish[:nrValidSamples], dataSetEnglish[nrValidSamples:]




#save the dataset in another file, in order to just load the other file and do not repeat this preprocessing and split

#Train
dict = {'portuguese': dataSetPortugueseTrain, 'english':dataSetEnglishTrain} 
df = pd.DataFrame(dict)
df.to_csv('portugueseEnglishTrain.csv', index=False)

#Validation
dict = {'portuguese': dataSetPortugueseValid, 'english':dataSetEnglishValid} 
df = pd.DataFrame(dict)
df.to_csv('portugueseEnglishValid.csv', index=False)
```
